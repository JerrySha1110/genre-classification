## Text PreProcessing

```python
#exploring what this data looks like
df.text.head()
```

```python
# I noticed that line breaks do not have a space after them. So when I drop line breaks I need to replace them with a space
#remove line breaks
df['text'] = df['text'].map(lambda s: re.sub(r' \n|\n', ' ', s))
```

```python
#examining text in round brackets
round_brackets = sum(list(df['text'].map(lambda s: re.findall(r'\((.*?)\)',s))), [])
#Number of round brackets:
print((len(round_brackets)))
```

```python
#Viewing some text in round brackets
#These just look like normal lyrics
random.seed(0)
random.choices(round_brackets, k=15)
```

```python
#examining text in square brackets
square_brackets = sum(list(df['text'].map(lambda s: re.findall(r'\[(.*?)\]',s))), [])
#how many instances of square brackets?
print((len(square_brackets)))
```

```python
#Viewing some text in square brackets
random.seed(0)
random.choices(square_brackets, k=15)

#these appear to be song features - so will drop these like we did in the previous dataset
```

```python
#re.sub(pattern, repl, string, count=0, flags=0)
# remove round brackets but not text within
df['text'] = df['text'].map(lambda s: re.sub(r'\(|\)', '', s))

# remove square brackest and text within
df['text'] = df['text'].map(lambda s: re.sub(r'\[(.*?)\] ', '', s))
```

```python
df.info()
```

## Removing Non-English Songs

Since the data was significantly more messy, our function that we had previously defined to detect english probability had a difficult time moving past characters it could not recognize. To deal with this issue I began by removing all non alphabetic characters, and dropped any row there the text data contianed, "te, que, el, comme, avec, sur, si, and wa", which are some of the most commonly used words in several popular languges. I dropped all songs with text that simply said "INSTRUMENTAL" and also used a word count to drop any song with 15 or less words. 

```python
#removing punctuation to get rid of weird characters that I noticed in the text
df["text"] = df['text'].str.replace('[^\w\s]','')
```

```python
df["text"]
```

```python
#Exploring songs that might be in non-english based on most frequently used words in other languages
df = df[~df['text'].str.contains(" te ")]
df = df[~df['text'].str.contains(" que ")]
df = df[~df['text'].str.contains(" el ")]
df = df[~df['text'].str.contains(" comme ")]
df = df[~df['text'].str.contains(" avec ")]
df = df[~df['text'].str.contains(" sur ")]
df = df[~df['text'].str.contains(" si ")]
df = df[~df['text'].str.contains(" wa ")]
#removing rows where there are no lyrics and the text body just says "INSTRUMENTAL"
df = df[df['text'] != "INSTRUMENTAL"]
```

```python
#finding the word count of each song
df['word_count'] = df['text'].str.split().str.len()
#dropping all songs with 15 words or less
df = df[df['word_count'] > 15]
df
```

```python
#Looking at the ros that don't include the word "a" to check for non-English songs remaining
df[~df['text'].str.contains(" the ")]
```

### Stemming and Preprocessing

```python
stemmer = PorterStemmer()
def lemmatize_stemming(text):
 
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

def preprocess(text):

    result = []
    #stopwords = list(gensim.parsing.preprocessing.STOPWORDS)
    #stopwords.pop(stopwords.index('no'))
    for token in gensim.utils.simple_preprocess(text):
        #if token not in set(stopwords):
        result.append(lemmatize_stemming(token))
    return ' '.join(result)
```

```python
df['text_preprocessed'] = df['text'].map(preprocess)
```

```python
df.head()
```

```python
#Checking what the tokens column look like before removing stopwords
df['text_preprocessed'].head()
```

## Tokenization

```python
#w+ for whitespace
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')

#creating a new column called tokens
df['tokens'] = df['text_preprocessed'].map(tokenizer.tokenize)
```

```python
df.head()
```

### Removing stop words
Reference for more info: https://www.geeksforgeeks.org/removing-stop-words-nltk-python/

```python
#setting stop words
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
```

```python
#removing stop words
df['tokens'] = df['tokens'].apply(lambda x: [item for item in x if item not in stop_words])
```

```python
df.head()
```

```python
#checking out the data
df.info()
```

```python
#checking out the distribution of the data
df.describe().transpose()
```

```python
#turning the stems back into untokenized structrure so that we can run count vector on it
df['stem_str'] = df['tokens'].map(lambda lst: ' '.join(lst))
```

```python
df.head()
```

### Vectorization

```python
df['tokens2']=df.tokens.apply(lambda x:str(x).strip('[]'))
df.head()                  
```

## Doc2Vec

```python
train_tokens_documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(df.tokens)]
```

```python
train_tokens_documents[1]
```

```python
model_holdout = gensim.models.doc2vec.Doc2Vec(train_tokens_documents,
                                            vector_size = 50, 
                                            min_count = 2, 
                                            epochs = 20) #Takes 5min
```

```python
model_holdout.train(train_tokens_documents, 
                  total_examples = model_holdout.corpus_count, 
                  epochs = model_holdout.epochs) #Takes 5min
```

```python
# used for running predictions on new data
model_holdout.infer_vector(train_tokens_documents[0].words, steps=20) #This changes every time. But within a certain range.
```

```python
def vector_for_learning(model, input_docs):
    sents = input_docs
    targets, feature_vectors = zip(*[(doc.tags[0], model.infer_vector(doc.words, steps=20)) for doc in sents])
    return targets, feature_vectors
```

```python
target_holdout, veclyrics_doc2vec_holdout = vector_for_learning(model_holdout, train_tokens_documents) #Takes 5min
```
